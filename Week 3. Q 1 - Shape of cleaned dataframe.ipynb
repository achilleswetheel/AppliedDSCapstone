```python
import pandas as pd
import numpy as np
import requests
from urllib.request import urlopen as uReq
from bs4 import BeautifulSoup
```

```python
#Getting the wikipedia page.

url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
html_doc=requests.get(url)
```

```python
#Using Beautiful soup to scrap the webpage and focus on the table.

soup = BeautifulSoup(html_doc.text, 'html.parser')
postaltable = soup.find('table', class_ = 'wikitable sortable').find_all('tbody')
 
for rows in postaltable:
    row=rows.find_all('td')

PostalCode = []
Borough=[]
Neighbourhood = []

count =3

#The table follows a pattern. 
# The 1,4,7,10... etc rows are the Borough values
# The 2,5,8,11  etc rows are the Neighbourhood values
# The 3,6,9,12... etc rows are the PostalCode values

for r in row:
    if(count%3 == 0):
        PostalCode.append(str(r.text.strip()))
    if(count%3 == 1):         
        Borough.append(str(r.text.strip()))
    if(count%3 == 2):         
        Neighbourhood.append(str(r.text.strip()))
    count = count+1
    
data = {'PostalCode':PostalCode, 
        'Borough':Borough,
        'Neighbourhood':Neighbourhood
       } 
  
#Thus we have got the values arranged in a dictionary and then we create DataFrame 
df = pd.DataFrame(data)

print(df.head(10))
    
    

```

```python
# We filter the dataframe to remove the 'Not assigned' rows.

df_filtered = df[df['Borough'] != 'Not assigned']
df_filtered = df_filtered.sort_values(by ='PostalCode')
df_filtered
```
